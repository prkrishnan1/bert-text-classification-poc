### Create AutoML Wrappers and start training job

```python
import sagemaker
from sagemaker import AutoML
from time import gmtime, strftime, sleep

timestamp_suffix = strftime("%d-%H-%M-%S", gmtime())
base_job_name = "automl-snow-triaging-" + timestamp_suffix
```

```python
sess = sagemaker.Session()
sagemaker_session_bucket=None
if sagemaker_session_bucket is None and sess is not None:
    # set to default bucket if a bucket name is not given
    sagemaker_session_bucket = sess.default_bucket()

role = sagemaker.get_execution_role()
sess = sagemaker.Session(default_bucket=sagemaker_session_bucket)
```

```python
target_attribute_name = "cmdb_ci"
target_attribute_values = np.unique(data[target_attribute_name])
problem_type = "MulticlassClassification"
job_objective = {"MetricName": "Accuracy"}
```

```python
automl = AutoML(
    role=role,
    target_attribute_name=target_attribute_name,
    base_job_name=base_job_name,
    sagemaker_session=sess,
    max_candidates=20,
    problem_type=problem_type,
    job_objective=job_objective
)
```

```python
train_file = "train.csv"
automl.fit(train_file, job_name=base_job_name, wait=True, logs=True)
```

```python
from sagemaker import AutoMLJob
automljob = AutoMLJob(sagemaker_session=sess, job_name=base_job_name, inputs=train_file)
```

```python
automljob.describe()
```

```python
automl.best_candidate(job_name=base_job_name)
```

```python
automl.list_candidates(job_name=base_job_name)
```

```python
from sagemaker.automl.candidate_estimator import CandidateEstimator

best_candidate = automl.best_candidate(job_name=base_job_name)
bestCandidateEstimator = CandidateEstimator(best_candidate, sagemaker_session=sess)
```

```python
model_name = "automl-snow-text-classification-model"
automl_model = automl.create_model(name=model_name, sagemaker_session=sess, candidate=bestCandidateEstimator, inference_response_keys=['predicted_label', 'probability'])
```

```python
from sagemaker.serializers import CSVSerializer
automl_endpoint_name = "automl-text-classifier-endpoint"
automl.deploy(initial_instance_count=1, instance_type="ml.m5.2xlarge", candidate=bestCandidateEstimator, endpoint_name=automl_endpoint_name, serializer=CSVSerializer())
```

```python
from sagemaker.predictor import Predictor
from sagemaker.deserializers import CSVDeserializer

automl_predictor = Predictor(endpoint_name=automl_endpoint_name, sagemaker_session=sess, serializer=CSVSerializer(), deserializer=CSVDeserializer())
```

```python
test_file = 'test.csv'
predictions = automl_predictor.predict(test_df.drop('cmdb_ci', axis=1).to_csv(index=False, header=False))
```

```python
def measure_coverage_accuracy(true_labels, predicted_labels_and_scores, eval_fn):
    thresholds = [0.70, 0.75, 0.80, 0.85, 0.90, 0.95]
    print("Threshold     |     Coverage     |     Accuracy    ")
    for threshold in thresholds:
        indexes = [i for i in range(len(predicted_labels_and_scores)) if float(predicted_labels_and_scores[i][1]) >= threshold]
        
        coverage = len(indexes) / 3700
        
        true_covered = []
        pred_covered = []
        
        for index in indexes:
            true_covered.append(true_labels[index])
            pred_covered.append(predicted_labels_and_scores[index][0])
        
        score = eval_fn(true_covered, pred_covered)
        
        print(threshold, "    |     ", coverage, "    |    ", score)
        
            
```

```python
def measure_coverage_ci(true_labels, predicted_labels, scores, eval_fn):
    unique_labels = true_labels.unique()
    coverage_ci = {}
    
    for label in unique_labels:
        y_true_bool = true_labels == label
        y_pred_bool = predicted_labels == label
        covered_scores = scores >= 0.75
        correct_preds = y_true_bool & y_pred_bool
        
        covered_and_correct = covered_scores & correct_preds
        
        
        label_accuracy = np.sum(covered_and_correct) / np.sum(y_true_bool)
        coverage_ci[str(label)] = label_accuracy
    
    return coverage_ci    
    
```

```python
from sklearn.metrics import accuracy_score

true_labels = test_df['cmdb_ci'].values
predicted_labels = np.array(predictions)

measure_coverage_accuracy(true_labels, predicted_labels, accuracy_score)
```
